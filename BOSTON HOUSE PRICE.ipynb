##1) Import Tensorflow

```python
import tensorflow as tf
print(tf.__version__)
```

##2) Load Boston house dataset

```python
from tensorflow.keras.datasets import boston_housing
```

```python
(features, prices), _ = boston_housing.load_data(test_split=0)
```

##3) DataSet Details

```python
print("Number of samples:", features.shape[0])
print("Number of features:", features.shape[1])
print("Target shape:", prices.shape)
```

##4) Build The Model

```python
# Initialize Sequential model
model = tf.keras.models.Sequential()

# Normalize input data
model.add(tf.keras.layers.BatchNormalization(input_shape=(13,)))

# Add Dense layer for prediction - Keras declares weights and bias automatically
model.add(tf.keras.layers.Dense(1))
```

##5) UserWarning Has Fixed

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Input

model = Sequential()
```

## - Input layer

```python
model.add(Input(shape=(13,)))
```

## - Normalize data

```python
model.add(BatchNormalization())
```

## - Output layer (Regression → 1 neuron)

```python
model.add(Dense(1))
```

##6) Compile and fit the Model

```python
model.compile(
    optimizer='sgd',
    loss='mse'
)
model.fit(
    features,
    prices,
    epochs=5,
    batch_size=32
)
```

##7) Make Prediction

```python
predicted_price = model.predict(features[:1])
print("Predicted Price:", predicted_price)
print("Actual Price:", prices[0])
```
